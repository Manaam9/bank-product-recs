## Оптимизация памяти

````python
# Выполним приведение типов данных, чтобы уменьшить объем требуемой памяти
def memory_usage_mb(df: pd.DataFrame) -> float:
    return df.memory_usage(deep=True).sum() / 1024**2


def optimize_dataframe_memory(
    df: pd.DataFrame,
    category_threshold: float = 0.5,
    exclude_cols: list[str] | None = None,
) -> pd.DataFrame:
    """
    Уменьшает потребление памяти DataFrame.
    
    category_threshold:
        если доля уникальных значений в object-колонке меньше этого порога,
        колонка переводится в category.
    exclude_cols:
        список колонок, которые не нужно трогать.
    """
    if exclude_cols is None:
        exclude_cols = []

    df_opt = df.copy()
    start_mem = memory_usage_mb(df_opt)
    print(f"Память до оптимизации: {start_mem:.2f} MB")

    for col in df_opt.columns:
        if col in exclude_cols:
            continue

        col_type = df_opt[col].dtype

        if pd.api.types.is_integer_dtype(col_type):
            c_min = df_opt[col].min()
            c_max = df_opt[col].max()

            if c_min >= 0:
                if c_max <= np.iinfo(np.uint8).max:
                    df_opt[col] = df_opt[col].astype(np.uint8)
                elif c_max <= np.iinfo(np.uint16).max:
                    df_opt[col] = df_opt[col].astype(np.uint16)
                elif c_max <= np.iinfo(np.uint32).max:
                    df_opt[col] = df_opt[col].astype(np.uint32)
                else:
                    df_opt[col] = df_opt[col].astype(np.uint64)
            else:
                if np.iinfo(np.int8).min <= c_min <= np.iinfo(np.int8).max and c_max <= np.iinfo(np.int8).max:
                    df_opt[col] = df_opt[col].astype(np.int8)
                elif np.iinfo(np.int16).min <= c_min <= np.iinfo(np.int16).max and c_max <= np.iinfo(np.int16).max:
                    df_opt[col] = df_opt[col].astype(np.int16)
                elif np.iinfo(np.int32).min <= c_min <= np.iinfo(np.int32).max and c_max <= np.iinfo(np.int32).max:
                    df_opt[col] = df_opt[col].astype(np.int32)
                else:
                    df_opt[col] = df_opt[col].astype(np.int64)

        elif pd.api.types.is_float_dtype(col_type):
            c_min = df_opt[col].min()
            c_max = df_opt[col].max()

            if np.finfo(np.float16).min <= c_min <= np.finfo(np.float16).max and np.finfo(np.float16).min <= c_max <= np.finfo(np.float16).max:
                df_opt[col] = df_opt[col].astype(np.float16)
            elif np.finfo(np.float32).min <= c_min <= np.finfo(np.float32).max and np.finfo(np.float32).min <= c_max <= np.finfo(np.float32).max:
                df_opt[col] = df_opt[col].astype(np.float32)
            else:
                df_opt[col] = df_opt[col].astype(np.float64)

        elif pd.api.types.is_object_dtype(col_type):
            num_unique = df_opt[col].nunique(dropna=False)
            num_total = len(df_opt[col])
            if num_total > 0 and (num_unique / num_total) < category_threshold:
                df_opt[col] = df_opt[col].astype("category")

    end_mem = memory_usage_mb(df_opt)
    print(f"Память после оптимизации: {end_mem:.2f} MB")
    print(f"Экономия: {start_mem - end_mem:.2f} MB")
    print(f"Снижение: {(start_mem - end_mem) / start_mem * 100:.2f}%")

    return df_opt
````

````python
# произведем приведение типов
# числовые признаки
numeric_object_cols = ["age", "antiguedad"]

for col in numeric_object_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# даты 
date_cols = ["fecha_dato", "fecha_alta", "ult_fec_cli_1t"]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# продуктовые колонки
product_cols = [c for c in df.columns if c.endswith("_ult1")]

for col in product_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype(np.uint8)

# бинарные признаки
binary_cols = ["ind_nuevo", "indrel", "tipodom", "ind_actividad_cliente"]

for col in binary_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
````

````python
def analyze_dataframe(df):
    summary = pd.DataFrame({
        "dtype": df.dtypes.astype(str),
        "missing_count": df.isna().sum(),
        "missing_percent": (df.isna().sum() / len(df)) * 100,
        "non_null_count": df.notna().sum(),
        "memory_MB": df.memory_usage(deep=True, index=False) / (1024 ** 2)
    })

    unique_values = {}

    for col in df.columns:
        series = df[col]

        if series.dtype == "object":
            non_null = series.dropna()

            if not non_null.empty and isinstance(non_null.iloc[0], (list, np.ndarray)):
                unique_values[col] = series.explode().dropna().nunique()
            else:
                unique_values[col] = series.nunique()
        else:
            unique_values[col] = series.nunique()

    summary["n_unique"] = pd.Series(unique_values)

    summary["missing_count"] = summary["missing_count"].astype("int64")
    summary["non_null_count"] = summary["non_null_count"].astype("int64")
    summary["n_unique"] = summary["n_unique"].astype("int64")
    summary["missing_percent"] = summary["missing_percent"].round(4)
    summary["memory_MB"] = summary["memory_MB"].round(4)

    return summary.sort_values("missing_count", ascending=False)
````

````python
def plot_distribution_plotly(
    df,
    column,
    percentile=0.99,
    bins=80,
    show_log_hist=True,
    show_kde="auto",
    kde_points=300,
    discrete_threshold=30
):
    """
    Строит интерактивную визуализацию распределения числового признака.

    Отображает:
    - гистограмму распределения
    - boxplot под гистограммой
    - вертикальные линии квартилей Q1, медианы и Q3
    - опционально KDE-кривую
    - дополнительную гистограмму в логарифмической шкале для long-tail распределений

    Parameters
    ----------
    df : pandas.DataFrame
        Датафрейм с данными.
    column : str
        Название числового признака.
    percentile : float, default=0.99
        Верхний процентиль для отсечения длинного хвоста на основном графике.
    bins : int, default=80
        Количество бинов для гистограммы.
    show_log_hist : bool, default=True
        Строить ли дополнительную гистограмму в логарифмической шкале.
    show_kde : {"auto", True, False}, default="auto"
        Показывать ли KDE-кривую.
        "auto" — включать только для непрерывных признаков.
    kde_points : int, default=300
        Количество точек для KDE-кривой.
    discrete_threshold : int, default=30
        Если число уникальных значений меньше или равно порогу,
        признак считается дискретным.

    Returns
    -------
    None
    """

    data = df[column].dropna()

    if data.empty:
        print(f"Признак {column} не содержит значений для визуализации.")
        return

    if not np.issubdtype(data.dtype, np.number):
        print(f"Признак {column} не является числовым.")
        return

    upper = data.quantile(percentile)
    filtered = data[data <= upper]

    if filtered.empty:
        print(f"После отсечения по percentile для признака {column} не осталось значений.")
        return

    q1 = filtered.quantile(0.25)
    q2 = filtered.quantile(0.50)
    q3 = filtered.quantile(0.75)

    unique_count = filtered.nunique()
    is_integer_like = np.all(np.isclose(filtered % 1, 0))
    is_discrete = unique_count <= discrete_threshold or is_integer_like

    if show_kde == "auto":
        use_kde = not is_discrete
    else:
        use_kde = bool(show_kde)

    fig = make_subplots(
        rows=2,
        cols=1,
        shared_xaxes=True,
        vertical_spacing=0.12,
        row_heights=[0.82, 0.18]
    )

    fig.add_trace(
        go.Histogram(
            x=filtered,
            nbinsx=bins,
            marker_color="royalblue",
            opacity=0.75,
            histnorm="probability density" if use_kde else None,
            hovertemplate=(
                f"{column}: " + "%{x}<br>"
                + ("Плотность: %{y}<extra></extra>" if use_kde
                   else "Количество наблюдений: %{y}<extra></extra>")
            )
        ),
        row=1,
        col=1
    )

    if use_kde and filtered.nunique() > 1:
        try:
            x_grid = np.linspace(filtered.min(), filtered.max(), kde_points)
            kde = gaussian_kde(filtered)
            y_kde = kde(x_grid)

            fig.add_trace(
                go.Scatter(
                    x=x_grid,
                    y=y_kde,
                    mode="lines",
                    name="KDE",
                    hovertemplate=(
                        f"{column}: " + "%{x}<br>"
                        "Плотность KDE: %{y:.6f}<extra></extra>"
                    )
                ),
                row=1,
                col=1
            )
        except Exception:
            use_kde = False

    fig.add_trace(
        go.Box(
            x=filtered,
            name="Boxplot",
            marker_color="royalblue",
            boxpoints=False,
            hovertemplate=f"{column}: " + "%{x}<extra></extra>"
        ),
        row=2,
        col=1
    )

    fig.add_vline(
        x=q1,
        line_dash="dash",
        line_color="green",
        annotation_text="Q1 (25%)",
        annotation_position="top left"
    )

    fig.add_vline(
        x=q2,
        line_dash="dash",
        line_color="red",
        annotation_text="Медиана",
        annotation_position="top"
    )

    fig.add_vline(
        x=q3,
        line_dash="dash",
        line_color="orange",
        annotation_text="Q3 (75%)",
        annotation_position="top right"
    )

    fig.update_layout(
        title=(
            f"Распределение признака: {column}"
            + (f" | KDE: {'ON' if use_kde else 'OFF'}" if show_kde != False else "")
        ),
        height=760,
        template="plotly_white",
        showlegend=use_kde,
        bargap=0.05,
        font=dict(size=12)
    )

    fig.update_xaxes(
        title_text=f"Значения признака {column}",
        row=2,
        col=1
    )

    fig.update_yaxes(
        title_text="Плотность" if use_kde else "Количество наблюдений",
        row=1,
        col=1
    )

    fig.update_yaxes(
        title_text="",
        showticklabels=False,
        row=2,
        col=1
    )

    fig.show()

    if show_log_hist:
        fig_log = go.Figure()

        fig_log.add_trace(
            go.Histogram(
                x=data,
                nbinsx=bins,
                marker_color="royalblue",
                hovertemplate=(
                    f"{column}: " + "%{x}<br>"
                    "Количество наблюдений: %{y}<extra></extra>"
                )
            )
        )

        fig_log.update_layout(
            title=f"Long-tail распределение признака: {column} (логарифмическая шкала)",
            template="plotly_white",
            height=500,
            bargap=0.05,
            xaxis_title=f"Значения признака {column}",
            yaxis_title="Количество наблюдений (логарифмическая шкала)",
            yaxis_type="log",
            showlegend=False,
            font=dict(size=12)
        )

        fig_log.show()
````

## Формирование таргета

Формула таргета:

target = product(t) - product(t-1)

````python
def build_next_product_targets(
    df: pd.DataFrame,
    user_col: str,
    date_col: str,
    product_cols: List[str],
    drop_first_record: bool = True
) -> Tuple[pd.DataFrame, List[str]]:
    """
    Формирует таргеты новых продуктов:
    target_product = 1, если продукта не было в предыдущем наблюдении,
    но он появился в текущем.

    Parameters
    ----------
    df : pd.DataFrame
        Исходная таблица клиентских состояний.
    user_col : str
        Колонка с id клиента.
    date_col : str
        Колонка с датой наблюдения.
    product_cols : List[str]
        Список продуктовых колонок.
    drop_first_record : bool, default=True
        Удалять ли первое наблюдение клиента, для которого нет истории.

    Returns
    -------
    df_out : pd.DataFrame
        Таблица с target-колонками.
    target_cols : List[str]
        Список сформированных target-колонок.
    """
    df_out = df.copy()
    df_out = df_out.sort_values([user_col, date_col]).reset_index(drop=True)

    prev_products = (
        df_out.groupby(user_col)[product_cols]
        .shift(1)
    )

    target_cols = []
    for col in product_cols:
        target_col = f"target_{col}"
        df_out[target_col] = (
            (df_out[col] == 1) &
            (prev_products[col] == 0)
        ).astype("Int8")
        target_cols.append(target_col)

    df_out["has_history"] = prev_products.notna().any(axis=1).astype("Int8")

    if drop_first_record:
        df_out = df_out[df_out["has_history"] == 1].copy()

    return df_out, target_cols
````

````python
# продукты
product_cols = [c for c in df.columns if c.endswith("_ult1")]

# сформируем таргеты
df_target, target_cols = build_next_product_targets(
    df=df,
    user_col="ncodpers",
    date_col="fecha_dato",
    product_cols=product_cols,
    drop_first_record=True
)

print(df_target.shape)
print(target_cols[:5])
````
